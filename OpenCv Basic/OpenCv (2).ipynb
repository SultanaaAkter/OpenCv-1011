```python
#Rotation of image based on rotation angle .Using warpAffine() function

(imageH,ImageW) = imagedata.shape[:2]
```

```python
centerX,centerY = (imageH/2,ImageW/2)
```

```python
#Use the cv2.rotate function: Built into OpenCV, but requires constructing a rotation matrix 
#and explicitly applying an affine warp, making the code more verbose.
#Use the imutils.rotate function: Part of my imutils library.
# Makes it possible to rotate an image with OpenCV in a single function call.
#Use the imutils.rotate_bound function: Also part of my imutils library.
# Ensures that no part of the image is cut off during rotation.


#cv2.getRotationMatrix2D() function is used to make the transformation matrix M 
#which will be used for rotating a image.
#cv2.getRotationMatrix2D(center, angle, scale)

#Parameters: center: Center of rotation
#angle(): Angle of Rotation. Angle is positive for anti-clockwise and negative for clockwise.
#scale: scaling factor which scales the image
#Return: 23 Rotation Matrix M

#This is a type of affine transformation. An affine transformation is transformation 
#which preserves lines and parallelism. These transformation matrix are taken by warpaffine() function 
#as parameter and the rotated image will be returned.


imageRotate = cv2.getRotationMatrix2D((centerX,centerY),90,1.0)
rotateNow = cv2.warpAffine(imagedata,imageRotate,(ImageW,imageH))
humanimageShow( " Rotate image",rotateNow)
```

```python
imageRotate = cv2.getRotationMatrix2D((centerX,centerY),180,1.0)
rotateNow = cv2.warpAffine(imagedata,imageRotate,(ImageW,imageH))
humanimageShow( " Rotate image",rotateNow)
```

```python
imageRotate = cv2.getRotationMatrix2D((centerX,centerY),270,1.0)
rotateNow = cv2.warpAffine(imagedata,imageRotate,(ImageW,imageH))
humanimageShow( " Rotate image",rotateNow)
```

```python
# step 2 : generate kernel 

#the main application of image gradients lies within edge detection.
#We use gradients for detecting edges in images, which allows us to find contours and 
#outlines of objects in images
#We use them as inputs for quantifying images through feature extraction  in fact, highly successful 
#and well-known image descriptors such as Histogram of Oriented Gradients and SIFT are built upon image 
#gradient representations
#Gradient images are even used to construct saliency maps, which highlight the subjects of an image

#From there well learn about Sobel and Scharr kernels,
# which are convolutional operators, allowing us to compute the image gradients automatically 
#using OpenCV and the cv2.Sobel function (we simply pass in a Scharr-specific argument to cv2.Sobel 
#to compute Scharr gradients).

#Computing image gradients is a pre-processing step to detecting edges in images.


```

```python
#scharr has two value.Scharr is work by corresponding of derivative of x and derivative of y.
#if we work by corresponding x then only change the value of x. same process for y.
#when the value of scharr is zero(0) or geater than (0) then we declare -1 otherwise we declare 1,2,3
#if we use less than(<) ,then we work with derivative of y.
#derivative of x and derivative of y are the value of Gradient.
#0=black color, and 1=white color.

kernelGenerate = -1 if argmumentImage["scharr"] > 0 else 3
```

```python
#Show gradient image in X axis and Y axis using Sobel
```

```python
gradientImageDataX=cv2.Sobel(grayimage ,ddepth = cv2.CV_32F, dx = 1, dy = 0,ksize = kernelGenerate)
gradientImageDataY=cv2.Sobel(grayimage ,ddepth = cv2.CV_32F, dx = 0, dy = 1,ksize = kernelGenerate)
humanimageShow("Gradient Image",gradientImageDataX)

#Now from below photo we can see that the human edge detechsion.by white color
# we can understand the structure  and outline of the  human
#Using this outline, we could then apply contours to extract the actual objects from the region 
#or quantify the shapes so we can identify them later.
```
